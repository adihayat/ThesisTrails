```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
c_drop_percent   = np.array([0.01,0.03,0.05,0.08,0.09])
low_grad_acc     = np.array([0.9012,0.8956,0.8696,0.7472,0.6708])
high_grad_acc    = np.array([0.8536,0.8088,0.78,0.6808, 0.6888])
low_grad_saving  = np.array([0.9817,0.9532,0.91464,0.8352,0.8126])
high_grad_saving = np.array([0.9957,0.9725,0.9273,0.8639,0.8491])
plt.plot(c_drop_percent,low_grad_acc,label='low_grads')
plt.plot(c_drop_percent,high_grad_acc,label='high_grads')
plt.legend()
plt.title("acc vs channel drop")

plt.figure()
plt.plot(1-low_grad_saving,low_grad_acc,label='low_grads')
plt.plot(1-high_grad_saving,high_grad_acc,label='high_grads')
plt.legend()
plt.title("acc vs weight saving")
```
